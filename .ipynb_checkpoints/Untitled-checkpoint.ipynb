## Building class


```python
number_of_rows = train.shape[0]

# distribution
(train[['Id','MSSubClass']].groupby('MSSubClass').count()/number_of_rows).sort_values(by='Id',ascending=0).plot(kind='bar')
(train[['Id','MSSubClass']].groupby('MSSubClass').count()/number_of_rows).plot(kind='bar')
```

```python
def boxplot_sorted(df, by, column, rot=0):
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    meds = df2.median().sort_values()
    return df2[meds.index].boxplot(rot=rot, return_type="axes")

boxplot_sorted(train, 'MSSubClass', 'SalePrice')
train.boxplot(column='SalePrice', by='MSSubClass')
```

This feature must be transformed one way or another. To avoid one-hot encoding, we may have to assume an order on the classes (based on mean of SalePrice?). This could induce the linear relationship showed above.

It may also make sense to group "similar" classes, although of course deciding how to group them may be an issue.

That said, that seems to be a Paretto when looking at class distribution. Classes 20 and 60 amount to more than 55% of data.
